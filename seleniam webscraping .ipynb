2.Please note that you have to scrape full job description. For that you may have to 
open each job separately as shown below

```python
job_info = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
job_info
```

```python
job_discp_url = []
for i in job_info:
    job_discp_url.append(i.get_attribute('href')) # to get job link for each jobs
job_discp_url
```

```python
job_titles_description = []
for i in job_discp_url:
    driver.get(i)  #---------> to get each link for job description
    try:
        job_titles_description = driver.find_elements_by_xpath("//h1[@class='jd-header-title']")
        job_titles_description.append(job_titles_description.text)
    except:
        job_titles_description.append("--")
        
job_titles_description # detail for 10 jobs
```
